# 막대 그래프
* 각 항목별 값의 크기나 도수분포표를 막대 모양으로 시각화 한 것
* 자료의 범주별 빈도를 요약해서 나타내는 그래프

```r
# 사례 : 두 약물에 따른 환자 반응
dose <- c(20,30,40,45,60)
drugA <- c(16,20,27,40,60)
drugB <- c(15,18,25,31,40)
```

```r
# barplot
# names.arg : 막대그래프에서 x축 각 이름 지정
# horiz : 수직/수평
barplot(drugA, xlab='pat', ylab='react Val', main='drugA react', col=c('navy','green','yellow','orange','red'),
       names.arg=1:5)
```

```r
barplot(drugA, xlab='pat', ylab='react Val', main='drugA react', col=c('navy','green','yellow','orange','red'),
       names.arg=1:5, horiz=T)
```

### 핫도그 먹기대회에서 최다 우승국순으로 시각화

```r
hotdog <- read.csv('./data/hotdog.csv')
```

```r
str(hotdog)
```

```r
wins <- sort(table(hotdog$Country), decreasing = T)
wins
```

```r
barplot(wins, col=c('red','orange','yellow','green'), xlab='country', ylab='wins', main='hotdog win country')
```

### 핫도그 먹기대회에서 최다 우승자순으로 시각화

```r
wins <- sort(table(hotdog$Winner), decreasing = T)
wins
```

```r
barplot(wins, col=c('red','orange','yellow','green'), xlab='winner', ylab='wins', main='hotdog winners')
```

### R에서 사용가능한 색상표
* RColorBrewer : R 내장 색상표

```r
library(RColorBrewer)
```

```r
display.brewer.all()
```

```r
# brewer.pal(n=색상수, 팔레트이름)
barplot(wins, col=brewer.pal(n=9, 'PuBu'), las=2, xlab='winner', ylab='wins', main='hotdog winners')
```

```r
# brewer.pal(n=색상수, 팔레트이름)
barplot(wins, col=rainbow(20), las=2, xlab='winner', ylab='wins', main='hotdog winners')
```
