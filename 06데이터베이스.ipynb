### 데이터 외부에 저장하기
* 메모리 내에 생성된 객체를 파일형태로 저장하고 싶다면
* write.csv, write.table, write_csv, fwrite등의 함수를 사용

```r
# row.names : 행번호 포함 여부
# quote : 인용부호(따옴표) 포함여부
write.csv(x=Nile, file='./data/Nile.csv', row.names=F, quote=F)
```

```r
write.csv(x=AirPassengers, file='./data/AirPassengers.csv', row.names=F, quote=F)
```

```r
write.csv(x=iris, file='./data/iris.csv', row.names=F, quote=F)
```

# MariaDB 다루기
* RMariaDB라는 패키지를 설치해야만 작업 가능

```r
url <- 'bigdata.cto1qluvf6xh.ap-northeast-2.rds.amazonaws.com'
uid <- 'admin'
pwd <- 'Bigdata_2023'
db <- 'bigdata'
```

```r
# 패키지 설치 : install.packages('패키지명')
install.packages('RMariaDB')
```

```r
# 패키지 불러오기 : libary(패키지명)
library(RMariaDB)
```

```r
conn <- dbConnect(
  drv = RMariaDB::MariaDB(),
  username = uid,
  password = pwd,
  host = url,
  dbname = db
)
```

```r
dbListTables(conn)
```

```r
dbDisconnect(conn)
```
