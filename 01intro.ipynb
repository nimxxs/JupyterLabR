# R
* R은 통계 계산과 그래픽을 위한 프로그래밍 언어이자 소프트웨어 환경이다
* R은 1960년대와 1970년대 Bell 연구소에서 개발된 S라는 데이터 처리 언어에 기반을 두고 있음
* 1990년대 중반 뉴질랜드 오클랜드 대학의 로스 이하카와 로버트 젠틀맨에 의해 시작되어
* 현재는 R의 핵심 기능은 R 코어 팀이, 다양한 추가 기능은 자발적 기여자들에 의해 개발되고 있음
* R은 GPL 하에 배포되는 공개 소프트웨어로 누구나 자유롭게 이용할 수 있음
* R은 빅데이터 분석에 널리 사용되고 있으며, 패키지 개발이 용이하여 통계 분석가들 사이에서 통계 소프트웨어 개발에 많이 쓰이고 있음

### R의 주요기능
* R은 데이터 조작, 계산, 그래픽 표현을 위한 소프트웨어
* 효율적이고 편련한 데이터 조작 및 처리 기능
* 데이터를 다양한 그래프로 표현해주는 데이터 시각화 기능
* 통계 분석 및 데이터 마이닝 알고리즘 수행 기능
* 분석 결과를 문서 및 발표 자료로 생성하는 기능
* 간단하며 효과적인 프로그래밍 언어로서의 기능

### R Studio
* R 배포판이 제공하는 기능만으로도 기본적인 프로그래밍이나 데이터 분석이 가능
* 그러나 프로그래밍이나 데이터 분석 작업이 복잡해지면 R을 좀 더 편리하게 사용할 수 있는
* 통합된 개발 환경(IDE)이 큰 도움이 됨
* 
RStudio는 R을 위한 강력한 통합 개발 환경을 제 줌
* R과 마찬가지로 공개 소프트웨어로 누구나 자유롭게 이용할 수 있음.


### R 맛보기

```r
# 결과는 tidyverse 형태로 출력
cars
```

```r
# 데이터 셋에 대한 도움말 확인
?cars
```

```r
summary(cars)
```

```r
# 산점도로 시각화
plot(cars)
```

```r
# 선형모델 생성 : lm(종속변수 ~ 독립변수, 데이터셋)
lm.cars <- lm(dist ~ speed, data=cars)
lm.cars
```

```r
# 선형모델 시각화
plot(cars)
abline(lm.cars, col='red')
```

#### 만약, 자동차의 속도가 3이라면 제동거리는?

```r
# 속도에 따른 제동거리 예측을 회귀분석 해보면
# 절편이 -17.579, 기울기가 3.932로 나옴
x <- 3
y <- 3.932 * x - 17.579
y
```
